```python
from bs4 import BeautifulSoup
import requests
import time

def link_philo(url):
    liens = []
    count = 0
    while count < 20 and not url.endswith("/wiki/Philosophie"):
        soup = BeautifulSoup(requests.get(url).text, 'html.parser')
        
        for balise_a in soup.find(id="mw-content-text").select('p a'):
            if balise_a.parent.name != 'span':
                hr = balise_a.attrs.get('href')
                
                if hr and hr.startswith('/wiki/'):
                    new_url = "https://fr.wikipedia.org"+hr
                    #liens.append(new_url)
                    if new_url != url:                
                        url = new_url
                    else :
                        url = url
                    print('found:', url)
                
        count += 1
        time.sleep(1)
    return count

url = "https://fr.wikipedia.org/wiki/Football"
test = link_philo(url)
#print(test)
```
