```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
pop_dir = "/home/parfait/Documents/MS BGD/INFMDI721_KIT_BigData/Cours1_2/INFMDI-721/lesson6/population_communes.csv"
dep_dir = "/home/parfait/Documents/MS BGD/INFMDI721_KIT_BigData/Cours1_2/INFMDI-721/lesson6/surface_departements.csv"

pop = pd.read_csv(pop_dir, sep = ',')
pop.head()
```

```python
dep = pd.read_csv(dep_dir, sep = ',')
dep.head()
```

```python
pop[ pop['variable'] == 'Population municipale' ]
```

```python
pop_mun = pop[pop['variable'] == 'Population municipale'].groupby(['Code département', 'Code région']).sum().reset_index()
```

```python
pop_mun.head()
```

```python
pop_num_fin = pop_mun.rename(columns={'value': 'Population municipale'})
```

```python
pop_tot = pop[pop['variable'] == 'Population totale'].groupby(['Code département', 'Code région']).sum().reset_index()
```

```python
pop_tot.head()
```

```python
pop_tot_fin = pop_tot.rename(columns={'value': 'Population totale'})
```

```python
pop_tot_fin.head()
```

```python
pop_new = pop_tot_fin.merge(pop_num_fin)
```

```python
pop_new.head()
```

```python
pop_new['pourcentage_pop_mun'] = (pop_new['Population municipale']/pop_new['Population totale']) * 100
```

```python
pop_new.head()
```

```python
pop_reg = pop.groupby(['Code région']).sum().reset_index()
```

```python
pop_reg_new = pop_reg.rename(columns={'value': 'Population region'})
```

```python
pop_reg_new.head()
```

```python
pop_final = pop_new.merge(pop_reg_new,  left_on=['Code région'], right_on=['Code région'])
```

```python
pop_final = pop_final.drop(['Code canton_x', 'Code arrondissement_x', 'Code commune_x', 'Code canton_y', 'Code arrondissement_y', 'Code commune_y' ], axis=1)
```

```python
pop_final.head()
```

```python
pop_by_reg = pop_final.merge(dep, left_on=['Code département'], right_on=['code_insee'])
```

```python
pop_by_reg.head()
```

```python
pop_by_reg['densite'] = pop_by_reg['Population municipale'] / pop_by_reg['surf_km2']
```

```python
pop_by_reg.head()
```
